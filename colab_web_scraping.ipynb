```python
from bs4 import BeautifulSoup as bs
import requests as rq
import re
```

*Capturando HTML*

```python
html_completo = rq.get("https://webscraping.andreregino.com.br/")
#html_completo = rq.get("https://www.atacadojandaia.com.br/")
html_formatado = bs(html_completo.content)
print(html_formatado.prettify())
```

```python
prinmeiro_h2 = html_formatado.find("h2").text
print(prinmeiro_h2)
```

```python
todas_ocorrencia_h2 = html_formatado.find_all("h2")
for ocorrencia_h2 in todas_ocorrencia_h2:
    print(ocorrencia_h2.text)
```

```python
h1_h2 = html_formatado.find_all(["h2","h1"])
for hh in h1_h2:
    print(hh.text)
```

```python
#preco = html_formatado.find_all(class_="card-title pricing-card-title")
preco = html_formatado.find(class_="card-title pricing-card-title")
#print(preco)
print(preco.text)
pre_requisito = html_formatado.find(id="pre-requisite-section")
print(pre_requisito.text)
```

```python
h1_h2_text = html_formatado.find_all(["h2","h1"], text="Reviews")
for correncia in h1_h2_text:
    print(correncia.text)
```

```python
h1_h2_text = html_formatado.find_all(["h2","h1"], text=re.compile("Para que aprender"))
for correncia in h1_h2_text:
    print(correncia)
```
