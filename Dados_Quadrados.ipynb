```python
import matplotlib.pyplot as plt

#teste = [-1,10,5,25,14,-1]
quadrados = [1,4,9,16,15]

fig, ax = plt.subplots()

ax.plot(quadrados, linewidth=3)
#ax.plot(teste, linewidth=3)

ax.set_title("Quadrados", fontsize=24)
ax.set_xlabel("Valor", fontsize=14)
ax.set_ylabel("Quadrados dos valores", fontsize=14)

ax.tick_params(axis='both', labelsize=14)

plt.show()
```
